###Money Change
Find the minimum number of coins needed to change the input value
(an integer) into coins with denominations 1, 5, and 10.

```python
import java.util.Scanner;

public class Change {
    private static int getChange(int m) {
        int acumulador = 0;
        int coins = 0;

        while(acumulador-m != 0 && acumulador+10 <= m){
            acumulador += 10;
            coins++;
        }
        while(acumulador-m != 0 && acumulador+5 <= m){
            acumulador += 5;
            coins++;
        }
        
        while(acumulador-m != 0 && acumulador+1 <= m){
            acumulador += 1;
            coins++;
        }

        return coins;
    }

    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        int m = scanner.nextInt();
        System.out.println(getChange(m));
        scanner.close();
    }
}
```

###Car Fueling

You are going to travel to another city that is located ð‘‘ miles away from your home city. Your car can travel
at most ð‘š miles on a full tank and you start with a full tank. Along your way, there are gas stations at
distances stop1
,stop2
, . . . ,stopð‘› from your home city. What is the minimum number of refills needed?

```python
#include <iostream>
#include <vector>

using std::cin;
using std::cout;
using std::vector;
using std::max;


int compute_min_refills(int dist, int tank, vector<int> & stops) {
    int sto = 0;
    int par = 0;
    int n = stops.size();
    int adv = 0;

    if((dist - stops[n-1])  > tank){
        //cout << "hay error \n";
        return -1;
    }

    while(adv < n){
        if(adv > 0 && (stops[adv] - stops[adv-1] > tank)){
            return -1;
        }
       
        if((stops[adv] - par) > tank){
            par = stops[adv-1];
            sto++;
        }
        adv++;
    }

    if((dist-par) > tank){
        par = stops[adv-1];
        sto++;
    }
    return sto;

}

int main() {
    int d = 0;
    cin >> d;
    int m = 0;
    cin >> m;
    int n = 0;
    cin >> n;

    vector<int> stops(n);
    for (size_t i = 0; i < n; ++i)
        cin >> stops.at(i);
    cout << compute_min_refills(d, m, stops) << "\n";
    //cout << compute_min_refills2(d, m, stops) << "\n";

    return 0;
}
```

Given two sequences ð‘Ž1, ð‘Ž2, . . . , ð‘Žð‘› (ð‘Žð‘–

the average number of clicks per day of the ð‘–-th slot), we need to partition them into ð‘› pairs (ð‘Žð‘–
, ð‘ð‘— )
such that the sum of their products is maximized.

```python
import java.util.*;

public class DotProduct {
    private static long maxDotProduct(int[] a, int[] b) {
        
        Arrays.sort(a);
        Arrays.sort(b);

        long result = 0;
        for (int i = 0; i < a.length; i++) {
            result = result+ (long)a[i] * b[i];
        }
        return result;
    }

    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        int n = scanner.nextInt();
        int[] a = new int[n];
        for (int i = 0; i < n; i++) {
            a[i] = scanner.nextInt();
        }
        int[] b = new int[n];
        for (int i = 0; i < n; i++) {
            b[i] = scanner.nextInt();
        }
        System.out.println(maxDotProduct(a, b));
        scanner.close();
    }
}

```

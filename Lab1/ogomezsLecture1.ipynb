##Euler formula:
The last function can be expressed as:

$$
  \sum_{i =0}^n=\frac{n(n+1)}{2}
$$




```
# Para que sea igual al resultado de arriba, es necesario restar 1, pues range es no inclusivo
def eulerSum(n):
  return ((n-1)*((n-1)+1))/2

eulerSum(10000000)
```

# Series arim√©ticas

Son las series en donde se se determina un valor a partir del anterior sumandole alguan constante. 

Al reducirlas mediante la indentidad de la suma, se pueden representar como:
$$
  \sum_{k=1}^n = \frac{1}{2}(n+1)
$$

Siendo esta la expresi√≥n de Gauss para la sumatoria de terminos.
Ahora en la forma m√°s general, se puede escribir como:
$$
  S=\frac{n[2a_1+(n-1)d]}{2}
$$
En donde $d$ es la constante y n el numero de veces que se ejecutara la sucesi√≥n.


## Algunas series 

### 1
$$ \sum_{k = 0}^n \frac{k^2(3*k)}{2(k*3)} , k > 0, k = 2p, p \in N \\
 $$

 Si se escoge $n = 1000$, el resultado es:

```
# Sumatoria de arriba

def sumatoria_uno(n):
    r = 0
    for k in range(1, n+1, 2):
      r = r + ((pow(k,2)*(3*k))/(2*(k*3)))
    return r     


sumatoria_uno(1000+1)
```

```
# Serie de fibonnacci
def fibonacci(n):
  if n == 0:
    serie = []
  elif n == 1:
    serie = [1]
  elif n == 2:
    serie = [0 , 1]
  elif n > 2:
    serie = [0, 1] 
    i = 1
    while i < n-1:
      serie.append(serie[i]+serie[i-1])
      i = i+1
  return serie     

fibonacci(151356512)
```

## Series geometricas

Son las series en las que la razon de cambio permanece constante, un ejemplo es:

$$
  \sum_{k = 1}^n = \frac{1}{2^k}
$$

Si $n = 20$ :

```
def serie_gem(n):
  r = 0
  for k in range(1,n+1):
    r = r + (1/pow(2,k))
  return r  

serie_gem(20)
```

## Series de potencias 

Son las series que usualmente aparecen en las series de Taylor, generalmente se escriben que son de la forma:
$$
  \sum_{n = 0}^\infty = c_n (x-a)^n
$$

####Ejemplo:
Se tiena la siguiente serie de potencias:

  $$
    \sum_{n =1}^{50} \frac{2^n}{n}(4x-8)^n
  $$
  Con $x = 10$

```
def seriePot(n):
  r = 0
  for k in range(1, n+1):
    r = r + (pow(2,k)/k)*pow(((4*10)-8), k)
  return r 

seriePot(50)   
```

###Fibonacci Number

Given an integer ùëõ, find the ùëõth Fibonacci number ùêπ_n

```python
#include <iostream>
#include <cassert>


int fibonacci_naive(int n) {
    if (n <= 1)
        return n;

    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2);
}

int fibonacci_fast(int n) {
    int p = -1;
    int q = 1;
    int cont = 0;
    while(cont <= n){
        int temp = p + q;  
        p = q;
        q = temp; 
        cont += 1;
    }
    return q;
}

void test_solution() {
    assert(fibonacci_fast(3) == 2);
    assert(fibonacci_fast(10) == 55);
    for (int n = 0; n < 20; ++n)
        assert(fibonacci_fast(n) == fibonacci_naive(n));
}

int main() {
    int n = 0;
    std::cin >> n ;

    //std::cout << fibonacci_naive(n) << '\n';
    //test_solution();
    std::cout << fibonacci_fast(n) << '\n';
    return 0;
}
```

### Last Digit of a Large Fibonacci Number

Given an integer ùëõ, find the last digit of the ùëõth Fibonacci number ùêπùëõ (that is, ùêπùëõ mod 10)

```python
#include <iostream>

int get_fibonacci_last_digit_naive(int n) {   // este es el rapido
    if (n <= 1)
        return n;

    int previous = 0;
    int current  = 1;

    for (int i = 0; i < n - 1; ++i) {
        int tmp_previous = previous;
        previous = current;
        current = (tmp_previous + current)%10;
    }

    return current ;
}

int main() {
    int n;
    std::cin >> n;
    int c = get_fibonacci_last_digit_naive(n);
    std::cout << c << '\n';
    }
```

### Greatest Common Divisor

Given two integers ùëé and ùëè, find their greatest common divisor

```python
#include <iostream>

int gcd_naive(int a, int b) {
    int x,y;
    if(a > b){
      x = a;
      y = b;
    }
    else{
      x = b;
      y = a;
    }

    int res = x % y;
    while ( res > 0 ){
      x = y;
      y = res;
      res = x % y;
      //std::cout << x << " " << y << "  "  << res << "\n"; 
    }
  return y;
}

int main() {
  int a, b;
  std::cin >> a >> b;
  std::cout << gcd_naive(a, b) << std::endl;
  return 0;
}
```

### Fibonacci Number Again

Given two integers ùëõ and ùëö, output ùêπùëõ mod ùëö (that is, the remainder of ùêπùëõ when divided by ùëö).

```python
#include <iostream>

long long get_fibonacci_huge_naive(long long n, long long m) {
    if (n <= 1)
        return n;

    long long previous = 0;
    long long current  = 1;

    for (long long i = 0; i < n - 1; ++i) {
        long long tmp_previous = previous;
        previous = current;
        current = tmp_previous + current;
    }
    std::cout << current << '\n';
    return current % m;
}

long pisano_period(long long f){
    long previous = 0;
    long current = 1;
    long remainder = 0;

    for(int i = 0; i < f*f ; i++){
        long temp = 0;
        temp = current;
        current = (previous + current) % f;
        previous = temp;
        if(previous == 0 && current == 1 ){
            remainder = i+1;
        }
    }
    return remainder;
}

long long fibonacci_fast(long long m, long remainder) {
    long long p = 0;
    long long q = 1;
    int count = 1;

    while(count < remainder ){
        int temp_q = q;
        q = (q + p) % m; 
        p = temp_q;
        count += 1;
    }
    return q;
}

long long get_fibonacci_huge_fast(long long n, long long m){
    if(n<= 1){
        return n;
    }
    long period = pisano_period(m);
    long remainder = n % period;

    long long f = fibonacci_fast(m, remainder);
    return f % m;
}

int main() {
    long long n, m;
    std::cin >> n >> m;
    std::cout << get_fibonacci_huge_fast(n, m) << '\n';
}
```

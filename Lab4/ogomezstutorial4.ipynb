## Numpy and basic Pandas

```python
# Primero se tiene que importar la libreria
import numpy as np
```

```python
# Primero se crear una lista y luego se convierte a un arreglo de nunpy

price_list = [143.73, 145.83, 143.68, 144.02, 143.5, 142.62]
price_array = np.array(price_list)
print(price_array, type(price_array))
```

```python
# Igualmente se pueden crear arreglos de varias dimensiones 

Ar = np.array([[1,3], [2,4]])
print(Ar, type(Ar))
```

```python
# Se pueden obtener las dimensiones de los arreglos con shape()

print(Ar.shape)
```

```python
#Igualmente se puede acceder a los elementos de los arreglos
print(Ar[0])
print(Ar[1])
```

```python
# y se pueden acceder por las columnas 
print('First column:', Ar[:,0])
print('Second column:', Ar[:,1])
```

## Funciones sobre arreglos

```python
# Se aplica la función sobre cada uno de los elementos del arreglo

np.log(price_array)
```

```python
# Otra de las funciones puede ser la media
np.mean(price_array)
```

```python
# La desviación estandar
print(np.std(price_array))
```

```python
# La sumatoria de los elementos del arreglo
print (np.sum(price_array))
```

```python
# Obtener el maximo de los elementos del arreglo
print(np.max(price_array))
```

## Uso de Pandas

```python
# Primero se importa la libreria 
import pandas as pd
```

```python
# Primero se convierte una lista de python en una serie de pandas

price = [143.73, 145.83, 143.68, 144.02, 143.5, 142.62]
s = pd.Series(price)
print (s)
```

```python
# Se le puede agregar indices a la serie, o modificarla 

s = pd.Series(price, index = ['a', 'b', 'c', 'd', 'e', 'f'])
print (s)
```

```python
# Se pueden volver a reasignas los indices
s.index = [6,5,4,3,2,1]
print (s)
```

```python
print(s[1:])
print(s[:-2])
```

```python
# O se pueden reasignar los valores
print(s[4])
s[4] = 0
print (s)
```

```python
# Se pueden asignar el nombre a una serie 

s = pd.Series(price, name = 'Apple Prices')
print (s)
print (s.name)
```

```python
# Para obtener la descripción de la serie
print (s.describe())
```

```python
# Se pueden asignar los indices

time_index = pd.date_range('2017-01-01', periods = len(s), freq = 'D')
print(time_index)
s.index = time_index
print (s)
```

```python
s.index = [6,5,4,3,2,1]
print (s)
print (s[1])
```

```python
# Se utiliza iloc para acceder a los elementos de las serie
print(s.iloc[1])
```

```python
# O tambien se puede acceder al elemento mediante el indice

s.index = time_index
print (s['2017-01-03'])
```

```python
# Y se puede acceder a un rando de valores

print (s['2017-01-02':'2017-01-05'])
```

```python
# Se puede acceder a los valores de las series mediante condicionales

print (s[s < np.mean(s)])
print (s[(s > np.mean(s)) & (s < np.mean(s) + 1.64*np.std(s))])
```

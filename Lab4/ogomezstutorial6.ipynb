## Rate of return, mean and variance


### Como evaluar la tasa de retorno
La de un periodo unico periodo unico puede ser evaluada de la siguiente forma:

$$
  r = \frac{P_t}{P_0}-1=\frac{P_t =P_0}{P_0}
$$

Donde $r$ es la tase de retorno.


```python
# Esto se puede expresar de la siguiente manera

import numpy as np
rate_return = 102.0/100 - 1
print(rate_return)
```

De la misma manera se puede calcular la tasa de retorno para varios periodos, lo que seria de la siguiente manera:

$$
  r = (1+r_1)+(1+r_1)-1
$$

De la misma manera se pueden seguir agregando terminos, es decir agregando periodos.

### Media arimetica

$$
  \mu = \frac{\sum_{i=1}^n x_i}{n}
$$

Se puede calcular la media arimetica mediante la función media

```python
print (np.mean(aapl.log_price))
```

### Varianza
$$
  \sigma^2=\frac{\sum_{i=1}^n (x_i - \mu)^2}{n}
$$

```python
# Esta varianza se puede calcular mediante el uso de la funcióon var

print (np.var(aapl.log_price))
```

### Desviación estandar

$$
  \mu = \sqrt{\sigma^2}=\sqrt{\frac{\sum_{i=1}^n (x_i - \mu)^2}{n}}
$$

```python
# La desviación estandar se calcula mediante la función std

print (np.std(aapl.log_price))
```

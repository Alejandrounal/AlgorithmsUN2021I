```python
# resample agrupa la infomación en pequeñas series, dependiendo de lo que se solicite
# Y ademas se puede pasara como parametro la media de las nuevas series geradas
by_month = aapl.resample('M').mean()
print(by_month)
```

```python
by_week = aapl.resample('W').mean()
print (by_week.head())
```

```python
three_day = aapl.resample('3D').mean()
two_week  = aapl.resample('2W').mean()
two_month = aapl.resample('2M').mean()
print(three_day)
print(two_week)
print(two_month)
```

```python
std = aapl.resample('W').std()    # standard deviation
max = aapl.resample('W').max()    # maximum value
min = aapl.resample('W').min()    # minimum value

# Tambien se pueden usar otro tipo de funciones
print(std)
print(max)
print(min)
```

```python
#Se puede usar una función lamda para indicar el ultimo elemento de la serie

last_day = aapl.resample('M').agg(lambda x: x[-1])
print (last_day)
```

```python
monthly_return = aapl.resample('M').agg(lambda x: x[-1]/x[1] - 1)
print (monthly_return)
```

```python
print (monthly_return.mean())
print (monthly_return.std())
print (monthly_return.max())
```

```python
# Diff sirve para calcular la diferencia entre elementos consecutivos
# y pct_change es para el calculo del porcentaje de cambio

print(last_day.diff())
print (last_day.pct_change())
```

```python
# Se pueden llenar Not a Number con algún valor que se desee
daily_return = last_day.pct_change()
print (daily_return.fillna(0))
```

```python
# Tambien se pueden llenar con un metodo preestablecido

daily_return = last_day.pct_change()
print (daily_return.fillna(method = 'bfill'))
```

```python
# Con drop na solamente se estan removiendo de la serie 

daily_return = last_day.pct_change().dropna()
print (daily_return)
```

## Utlizando DataFrames

```python
# Se puede crear un data frame desde pasandole un diccionario

dict = {'AAPL': [143.5,  144.09, 142.73, 144.18, 143.77],
        'GOOG': [898.7,  911.71, 906.69, 918.59, 926.99],
        'IBM':  [155.58, 153.67, 152.36, 152.94, 153.49]}
dates = pd.date_range('2017-07-03', periods = 5, freq = 'D')
df = pd.DataFrame(dict, index = dates)
print(df)
```

```python
# De esta manera se pueden manipular los dataframes 

df = aapl_table
print (df.Close.tail(5))
print (df['Adj. Volume'].tail(5))
```

```python
aapl_2016 = df['2016']
aapl_month = aapl_2016.resample('M').agg(lambda x: x[-1])
print (aapl_month)
```

```python
# Igualmente se pueden acceder a los atributos por los nombres

aapl_bar = aapl_month[['Open', 'High', 'Low', 'Close']]
print(aapl_bar)
```

```python
# Se pueden especificar columnas con loc

print (aapl_month.loc['2016-03':'2016-06', ['Open', 'High', 'Low', 'Close']])
```

```python
# Se puede crear un subset apartir de un dataframe existente 

aapl_bar['rate_return'] = aapl_bar.Close.pct_change()
print(aapl_bar)
```

```python
# Se puede usar el metodo isnull para determinar como afectan los valores anormales en la serie

missing = aapl_bar.isnull()
print (missing)
print ('---------------------------------------------')
print (missing.describe())
```

```python
# Tambien se puede saber cuales filas tienen valores nulos

print(missing[missing.rate_return == True])
```

```python
# Los valores faltantes se puden descartar o llenar, igual que en las series

drop = aapl_bar.dropna()
print (drop)
print ('\n--------------------------------------------------\n')
fill = aapl_bar.fillna(0)
print (fill)
```

```python
# Es posible concatenar los dataframes con concat

s1 = pd.Series([143.5, 144.09, 142.73, 144.18, 143.77], name = 'AAPL')
s2 = pd.Series([898.7, 911.71, 906.69, 918.59, 926.99], name = 'GOOG')
data_frame = pd.concat([s1, s2], axis = 1)
print (data_frame)
```

```python
import numpy as np

# Axis = 1 permite que se concatenen los dataframes por columnas

log_price = np.log(aapl_bar.Close)
log_price.name = 'log_price'
print (log_price)
print ('\n--------------------------------------------\n')
concat = pd.concat([aapl_bar, log_price], axis = 1)
print (concat)
```

```python
# Tambien se pueden unir por filas

df_volume = aapl_table.loc['2016-10':'2017-04', ['Volume', 'Split Ratio']].resample('M').agg(lambda x: x[-1])
print (df_volume)
print ('\n-------------------------------------------\n')
df_2017 = aapl_table.loc['2016-10':'2017-04', ['Open', 'High', 'Low', 'Close']].resample('M').agg(lambda x: x[-1])
print (df_2017)
```

```python
concat = pd.concat([aapl_bar, df_volume], axis = 1)
print(concat)
```

```python
# Tambien se puede unir medienta la intersección  para esto se  utiliza el atributo inner

concat = pd.concat([aapl_bar, df_volume], axis = 1, join = 'inner')
print (concat)
```

```python
append = aapl_bar.append(df_2017)
print (append)
```

```python
# Se pueden unir los dataframe con axis 0, y esto reemplaza append
concat = pd.concat([aapl_bar, df_2017], axis = 0)
print(concat)
```

```python
# Si las columnas de los dataframes a ser unidos se llaman igual, estas simplemente se unen

df_2017.columns = ['Change', 'High', 'Low', 'Close']
concat = pd.concat([aapl_bar, df_2017], axis = 0)
print (concat)
```

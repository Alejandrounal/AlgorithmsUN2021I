```python
#   Se puedet= agregar un key como parametro a lo que se ira a imprimir, 
#   Lo que se esta creando es una lista de tuplas
price_list = [('AAPL', 144.09), ('GOOGL', 911.71), ('MSFT', 69), ('FB', 150), ('WMT', 75.32)]
sorted(price_list, key = lambda x: x[1])
```

```python
# Se puede cambiarel orden en el que se ordenaran los elementos

price_list = [('AAPL', 144.09), ('GOOGL', 911.71), ('MSFT', 69), ('FB', 150), ('WMT', 75.32)]
sorted(price_list, key = lambda x: x[1], reverse = True)
```

```python
# Utilizando directamente la funcion de ordenar del objeto list

price_list = [('AAPL', 144.09), ('GOOGL', 911.71), ('MSFT', 69), ('FB', 150), ('WMT', 75.32)]
price_list.sort(key = lambda x: x[1])
print(price_list)
```

##POO

```python
  #Creación de una clase
  class Stock:
    # Constructor de la clase
    def __init__(self, ticker, open, close, volume):
        self.ticker = ticker
        self.open = open
        self.close = close
        self.volume = volume
        self.rate_return = float(close)/open - 1

    # Metodo para la actualizaciónd de datos 
    def update(self, open, close):
        self.open = open
        self.close = close
        self.rate_return = float(self.close)/self.open - 1
    
    #Metodoq que devuelve la 
    def print_return(self):
        print (self.rate_return)
```

```python
# Creación de objetos de la clase
apple  = Stock('AAPL', 143.69, 144.09, 20109375)
google = Stock('GOOGL', 898.7, 911.7, 1561616)
```

```python
# LLamar los atributos del objeto creado
apple.ticker
```

```python
# Ahora llamar el metodo de la clase
google.print_return()
```

```python
# Se estan actualizando los atributos atravez de los metodos

google.update(912.8,913.4)
google.print_return()
```

```python
# Tambien se puede actualizar lo datos directamente
apple.ceo = 'Tim Cook'
apple.ceo
```

```python
# Sirve para conocer los atributos declarados para el objeto que se pasa como argumento

dir(apple)
```

## Herencia 


```python
# La clase de la que se hereda se pasa como si se tratara de un paametro

class Child(Stock):
    def __init__(self, name):
        self.name = name
```

```python
aa = Child('AA')
print(aa.name)
```

```python
# Se tienen los mismos metodos de la clase padre
aa.update(100, 102)
print(aa.open)
print(aa.close)
print(aa.print_return())
```
